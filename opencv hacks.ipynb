```python
import cv2
import numpy as np

img = cv2.imread("car2.jpg")

kernel = np.ones((4,4), np.uint8)

imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgBlur = cv2.GaussianBlur(imgGray, (7,7), 0)

imgCanny = cv2.Canny(img, 150, 150) # high threshold low edges
imgDialation = cv2.dilate(imgCanny, kernel, iterations = 1) # edge thicker
imgEroded = cv2.erode(imgDialation, kernel, iterations = 1) # edge thinner

imgCanny = cv2.resize(imgCanny, (480, 360))
imgDialation = cv2.resize(imgDialation, (480, 360))
imgEroded = cv2.resize(imgEroded, (480, 360))

cv2.imshow("Gray", imgGray)
cv2.imshow("Blur", imgBlur)

cv2.imshow("Edges", imgCanny)
cv2.imshow("Eroded", imgEroded)
cv2.imshow("Dialation", imgDialation)

cv2.waitKey(0) # msec

cv2.destroyAllWindows()
```

```python
import cv2

# warp prespective

# extract a card from multiple cards

img = cv2.imread("card.jpg")
img = cv2.resize(img, (500,500))

width, height = 240, 350 # playing card is of 2.4 * 3.5

pts = np.float32([[11,101],[30,299],[150,87],[172,280]])
pts_corner = np.float32([[0,0],[0,height],[width,0],[width,height]])

matrix = cv2.getPerspectiveTransform(pts, pts_corner)
imgout = cv2.warpPerspective(img, matrix, (width, height))

bg = np.zeros((512,512, 3), np.uint8)

for i in range(imgout.shape[0]):
    for j in range(imgout.shape[1]):
        bg[i+75][j+130] = imgout[i][j]
        
print(imgout.shape)

cv2.imshow("Cards",img)
cv2.imshow("Card", bg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# detect color

def empty(x):
    pass

import cv2
import numpy as np

cv2.namedWindow("Trackbars")
cv2.resizeWindow("Trackbars", 480,240)
cv2.createTrackbar("HueMin", "Trackbars", 67, 255, empty)
cv2.createTrackbar("HueMax", "Trackbars", 122, 255, empty)
cv2.createTrackbar("SatMin", "Trackbars", 0, 255, empty)
cv2.createTrackbar("SatMax", "Trackbars", 36, 255, empty)
cv2.createTrackbar("ValMin", "Trackbars", 88, 255, empty)
cv2.createTrackbar("ValMax", "Trackbars", 255, 255, empty)

img = cv2.imread("car2.jpg")
img = cv2.resize(img, (400, 400))

imghsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

while True:
    
    h_min = cv2.getTrackbarPos("HueMin", "Trackbars")
    h_max = cv2.getTrackbarPos("HueMax", "Trackbars")
    s_min = cv2.getTrackbarPos("SatMin", "Trackbars")
    s_max = cv2.getTrackbarPos("SatMax", "Trackbars")
    v_min = cv2.getTrackbarPos("ValMin", "Trackbars")
    v_max = cv2.getTrackbarPos("ValMax", "Trackbars")
    
    lower = np.array([h_min, s_min, v_min])
    upper = np.array([h_max, s_max, v_max])
    
    mask = cv2.inRange(imghsv, lower, upper)
    imgres = cv2.bitwise_and(img, img, mask = mask)
    
    maskbgr = cv2.cvtColor(mask,cv2.COLOR_GRAY2BGR)
    
    hor1 = np.hstack((img,imghsv))
    hor2 = np.hstack((imgres, maskbgr))
    images = np.vstack((hor1,hor2))
    images = cv2.resize(images, (600,600))
    
    cv2.imshow("All Images", images)
    
    if (cv2.waitKey(1) & 0xFF) == ord('q'):
        cv2.destroyAllWindows()
        break;
```

```python
# contours and shapes

import cv2
import numpy as np

def getcontours(img):
    # retrieve outer contours
    contours, hierarchy = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    for cnt in contours:
        area = cv2.contourArea(cnt)
#         print(area)
        if area > 500:
            cv2.drawContours(imgcontour, cnt, -1, (255,0,0), 3)
            peri = cv2.arcLength(cnt, True) # closed contour: True
#             print(peri)
            epsilon = 0.03 * peri # which is maximum distance from contour to approximated contour
            approx = cv2.approxPolyDP(cnt,epsilon,True)
         
            objcor = len(approx)
#             print(objcor)
            
            x,y,w,h = cv2.boundingRect(approx)

            if objcor == 3:
                objecttype = "Triangle"
            elif objcor == 4:
                aspectratio = w/float(h)
                if aspectratio > 0.95 and aspectratio < 1.05:
                    objecttype = "Square"
                else:
                    objecttype = "Rectangle"
            elif objcor == 5:
                objecttype = "Pentagon"                    
            else:
                objecttype = "Circle"
            
            cv2.rectangle(imgcontour, (x,y),(x+w,y+h),(0,255,0),2)
            cv2.putText(imgcontour, objecttype, (x, y + h + 20),
                       cv2.FONT_HERSHEY_COMPLEX, 0.7, (0, 0, 255), 1)

            
img = cv2.imread("shape1.png")
imgcontour = img.copy()

imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgblur = cv2.GaussianBlur(imggray,(7,7),1)
imgcanny = cv2.Canny(imgblur,50,50)

getcontours(imgcanny)

null = np.zeros(img.shape, np.uint8)

hor1 = np.hstack((imggray,imgblur,imgcanny))
hor1 = cv2.cvtColor(hor1,cv2.COLOR_GRAY2BGR)
hor2 = np.hstack((null, img, imgcontour))

images = np.vstack((hor1,hor2))
images = cv2.resize(images,(1200,600))
cv2.imshow("shapes", images);

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# face detection
# viola and jones method
# positive faces and negative non faces to train model
# save into casade file(XML)
# we are using pre-trained cascade file

import cv2
# fast but not accurate
facecascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

img = cv2.imread("team1.jpg")
img = cv2.resize(img,(1000,600))
imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = facecascade.detectMultiScale(imggray,2,5)

for (x,y,w,h) in faces:
    cv2.rectangle(img, (x-5,y-30),(x+w+5,y+h+5), (255,0,0), 3)

cv2.imshow("image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
